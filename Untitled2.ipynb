```python
df['booking_date'] = pd.to_datetime(df['booking_date'])

df['check_in_date'] = pd.to_datetime(df['check_in_date'],errors='coerce')

df['check_out_date'] = pd.to_datetime(df['check_out_date'],errors='coerce')
```

```python
df['lead_time'] = (df['check_in_date'] - df['booking_date']).dt.days
```

```python
print(df[['lead_time','length_of_stay','check_in_date','check_out_date']].isnull().sum())
```

```python
df['lead_time'] = df['lead_time'].fillna(df['lead_time'].median())

df['length_of_stay'] = df['length_of_stay'].fillna(df['length_of_stay'].median())
```

```python
df = df[
    df['booking_status'].isin(
        ['Confirmed', 'Cancelled']
    )
]

print(df['booking_status'].value_counts())
```

```python
df['booking_status'] = df['booking_status'].map({
    'Confirmed': 0,
    'Cancelled': 1
})

df['booking_status'].value_counts()
```

```python
df['booking_status'].value_counts()
```

```python
drop_columns = ['customer_id','property_id','refund_status','refund_amount','coupon_redeem']

df.drop(columns=drop_columns,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_columns = ['city','room_type','stay_type','booking_channel','payment_method','channel_of_booking','Coupon USed?']

for col in categorical_columns:
    df[col] = le.fit_transform(df[col].astype(str))
    
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
# Remove date columns permanently

df = df.drop(
    columns=[
        'booking_date',
        'check_in_date',
        'check_out_date',
        'travel_date'
    ],
    errors='ignore'
)

# Create fresh X and y

X = df.drop('booking_status', axis=1)
y = df['booking_status']

# Convert bool column to int

X['is_weekend_checkin'] = X['is_weekend_checkin'].astype(int)

print(X.dtypes)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.20,
    random_state=42)


```

```python
X_train
X_test
```

```python
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()
```

```python
model.add(Dense(units=128, activation="relu", input_shape=(X_train.shape[1],)))
model.add(Dropout(0.3))

model.add(Dense(units=128, activation="relu"))
model.add(Dropout(0.3))

model.add(Dense(units=128, activation="relu"))
model.add(Dropout(0.3))

model.add(Dense(units=64, activation="relu"))
model.add(Dropout(0.3))

model.add(Dense(units=64, activation="relu"))
model.add(Dropout(0.3))

model.add(Dense(units=64, activation="relu"))
model.add(Dropout(0.3))

model.add(Dense(units=32, activation="relu"))
model.add(Dropout(0.3))

model.add(Dense(units=32, activation="relu"))
model.add(Dropout(0.3))

model.add(Dense(units=32, activation="relu"))
model.add(Dropout(0.3))

model.add(Dense(units=32, activation="relu"))
model.add(Dropout(0.3))

model.add(Dense(units=1,activation="sigmoid"))

```

```python
model.summary()
```

```python
early=EarlyStopping(monitor="val_loss",verbose=1,patience=5)
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(X_train,y_train,epochs=50,batch_size=32,verbose=1,callbacks=[early],validation_data=(X_test,y_test))
```

```python
from sklearn.metrics import (confusion_matrix,classification_report,accuracy_score)
```

```python
y_pred_train=model.predict(X_train)
```

```python
y_pred=model.predict(X_test)
```

```python
loss, accuracy = model.evaluate(
    X_test,
    y_test,
    verbose=0
)

print("Test Accuracy:", accuracy)
print("Loss:", loss)
```

```python
y_prob = model.predict(X_test)

y_pred = (
    y_prob > 0.5
).astype(int)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(
    y_test,
    y_pred
)

print(cm)
```

```python
plt.figure(figsize=(6,4))

sns.heatmap(cm,
    annot=True,
    fmt='d',
    cmap='Blues'
)

plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```
